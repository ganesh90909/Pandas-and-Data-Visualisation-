#### Histogram and Density plots

```python
#density plots or, more formally, Kernel Density Plots.
#They can be considered a smoothed version of the histogram. 
#Their main advantage over the latter is that they do not depend on the size of the bins.
#Let's create density plots for the same two variables:

df[features].plot(kind='density',subplots=True,layout=(1,2),sharex=False,figsize=(10,4));
```

### t SNE(t-distributed Stochastic Neighbor Embedding)

it is used for representing d-dimensional data in a 2-d dimensional plane or sometimes in 3-d plane.

```python
from sklearn.manifold import TSNE
```

```python
df.head()
df.drop('State',axis=1,inplace=True)
```

```python
df.head()
```

```python
tsne=TSNE(random_state=17)
```

```python
%%time
X_repr=tsne.fit_transform(df)
```

```python
X_repr.shape
```

```python
plt.rcParams['figure.figsize']=(8,6)
plt.scatter(X_repr[df['Churn']==1,0],X_repr[df['Churn']==1,1],alpha=.7,c='blue',label='churned')
plt.scatter(X_repr[df['Churn']==0,0],X_repr[df['Churn']==0,1],alpha=.3,c='green',label='loyal')
plt.legend()
plt.xlabel('tSNE axis #1')
plt.ylabel('tSNE axis #2')
plt.title('tSNE representation')
plt.savefig('churn_tSNE.png',dpi=300)
```
