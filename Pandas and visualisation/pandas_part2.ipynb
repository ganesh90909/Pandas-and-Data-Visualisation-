### Grouping

#### In general grouping can be done by df.groupby(by=grouping-columns)[column-to-show].function()

```python
columns_to_show = ['Total day minutes', 'Total eve minutes', 
                   'Total night minutes']

df.groupby(['Churn'])[columns_to_show].agg([np.mean, np.std, np.min, 
                                            np.max])
```

### Pivot tables

Pivot tables are implementedd in pandas. pivot tables takes three parameters
  <ul><li>**values**-list of variables to calculate statistics for</li>
    <li>**index**-list of variables to group data by</li>
  <li>**aggfunc**-what statistics we need to calculate for groups, ex. sum, mean, maximum, minimum or something else</li>

   

```python
df.pivot_table(['Total day calls','Total eve calls','Total night calls'],['Area code'],aggfunc='mean')

```

## Dataframe Transformations

### First attempt at predicting churn telecom

Let's see how churn rate is related to the International plan feature. 
We'll do this using a crosstab contingency table and also through visual analysis with Seaborn.
                             

```python
pd.crosstab(df['Churn'],df['Customer service calls'],margins=False)
```

```python
sns.countplot('Customer service calls',hue='Churn',data=df)
```

```python
df['Many service calls']=(df['Customer service calls']>3).astype('int')
pd.crosstab(df['Many service calls'],df['Churn'],margins=True)  #'All' attribute is added by making margins True
```

```python
sns.countplot(x=df['Many service calls'],hue='Churn',data=df);
```

```python
pd.crosstab(df['Many service calls'],df['International plan'],margins=False)
```

```python
sns.countplot(x=df['Many service calls'],hue='International plan',data=df)
```

```python
pd.crosstab(df['Many service calls']&df['International plan'],df['Churn'])
```

```python
sns.countplot(x=(df['Many service calls']&df['International plan']),hue='Churn',data=df)
```
